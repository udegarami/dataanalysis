#### **Beispiel**: Rekonstruktion der Testdaten unter Verwendung der entsprechenden Trainingsdaten und des vollständigen Datensatzes

* **Full set**: https://www.tensorflow.org/datasets/catalog/titanic
* **Train set**: https://github.com/datasciencedojo/datasets/blob/master/titanic.csv (_\"\" durch \' ersetzt, un alle Datensäatye korrekt abzugleichen_)

```python
dfTotal = pd.read_csv('titanicFull.csv')
dfTrain = pd.read_csv('titanicTrain.csv')
print(f"Total records {len(dfTotal)}, Training Set records {len(dfTrain)}")
```

Verwendung von `.isin(<collection>)` zur Prüfung auf Überschneidungen/Komplemente 

```python
trainSetMatches = dfTotal["Name"].isin(dfTrain["Name"])
trainSetMatches.value_counts() # True muss 891 sein
```

Sobald die Übereinstimmungen 100% erreichen (891 mögliche Fälle), können die anderen Werte als Testsatz gespeichert werden.  
Wenn dies nicht der Fall ist, muss eine weitere Datenanalyse durch Slicing durchgeführt werden, um herauszufinden, was die Ursache für die fehlenden Übereinstimmungen ist. z. B. ergab die Überprüfung des Komplements von `dfTotal` in `dfTrain`, dass die fehlenden Fälle auf unterschiedliche Anführungszeichen für Spitznamen zurückzuführen waren

```python
dfTest = dfTotal[~trainSetMatches].copy()
dfTest.to_csv("titanicTest_generated.csv", index=False)
```

#### **Beispiel**: Primer in Merkmalsanalyse

Dieser Datensatz wird häufig verwendet, um die Wahrscheinlichkeit zu modellieren, dass jemand die Titanic-Katastrophe auf der Grundlage seiner Merkmale überlebt.  
Schauen wir uns die Daten an, die wir derzeit haben:

Der Titanic-Datensatz enthält eine Liste der bekannten Passagiere der Titanic mit den folgenden Variablen

* pclass: _Passagierklasse (1 = 1; 2 = 2; 3 = 3)_
* survival: _(0 = Nein; 1 = Ja)_
* sibsp: _Anzahl der Geschwister/Ehegatten an Bord (siblings-spouse auf Englisch)_
* parch: _Anzahl der Eltern/Kinder an Bord (parents-children auf Englisch)_
* embarked: _Einschiffungshafen (C = Cherbourg; Q = Queenstown; S = Southampton)_
* boat: _Nummer des Rettungsbootes, in dem die Person gezählt wurde_
* body: _Körperidentifikationsnummer_

```python
df = pd.read_csv('titanicFull.csv')
df.info()
```

Es hat keine Nullwerte! Wir haben jedoch zuvor gesehen, dass anstelle von Nullen „?“ verwendet wird, wenn keine Daten vorhanden sind

```python
df = df.replace("?", pd.NA)
df.info()
```

Aus den Variablen können wir ersehen, dass `Cabin`, `Boat`, `Body` und `HomeDest` ziemlich leer sind und für die weitere Analyse nicht hilfreich sind  
Diese Variablen könnten zwar leicht entfernt werden, aber beachten Sie den Fall von `Boat` und `Body`: Wenn jemand mit einem Rettungsboot verbunden war oder eine Körperidentifikationsnummer hat, ist es einfach, auf sein Schicksal zu schließen.

```python
df2= df[["Body","Boat","Survived"]].copy()
df2.head()
```

Diese können formaler interpretiert werden, indem zusätzliche Variablen erstellt werden, um festzustellen, ob das Attribut existiert oder nicht  
Während diese Variablen der Interpretation durch den Benutzer überlassen werden können, ist es eine gute Konvention, stattdessen Zahlen zu verwenden (1 = Ja, 0 = Nein)

```python
# Attributnamen können beliebig sein, müssen aber häufig eingegeben werden
df2["BodyQ"] = 0
df2.loc[~df2["Body"].isna(), "BodyQ"] = 1

df2["BoatQ"] = 0
df2.loc[~df2["Boat"].isna(),"BoatQ"] = 1
```

Die Ergebnisse können dann gruppiert und ausgezählt werden

```python
# oder df2.value_counts(["BodyQ","Survived"], sort=False)
df2[["BodyQ","Survived"]].groupby(["BodyQ"]).value_counts()
```

```python
# oder df2.value_counts(["BoatQ","Survived"], sort=False)
df2[["BoatQ","Survived"]].groupby(["BoatQ"]).value_counts()
```

Wenn die Attribute numerisch sind, können wir andere Berechnungen durchführen, z. B. Korrelationswerte

```python
df2[["BodyQ","BoatQ","Survived"]].corr()
```

Diese beiden Variablen erklären das Zielattribut signifikant. Da das Ziel jedoch darin besteht, mit Hilfe von ML-Algorithmen und -Techniken tiefer zu graben, werden sie oft weggelassen.  
Beachten Sie, dass wir auch bei direkter Verwendung dieser Variablen keine exakten Ergebnisse erhalten, sondern nur eine sehr vernünftige Schätzung.

### Matplotlib

Eine Bibliothek, die für Graphen und Diagramme verwendet wird. Pandas bietet Komfortfunktionen zur Nutzung seiner Funktionen (siehe https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```python
# Auch wenn das meiste über Pandas abgewickelt wird, müssen einige Optionen direkt geladen werden.
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanicFull.csv')
df = df.replace("?", pd.NA)
df.drop(columns=["Cabin","Boat","Body","HomeDest"])
```

Die übrigen Variablen können Aufschluss über die Verteilung der Fahrgäste geben. Zu diesem Zweck ist vielleicht ein grafischer Ansatz besser geeignet als das bloße Lesen von Zahlenlisten.  
Wir können auch Variablen identifizieren, die als diskrete Werte arbeiten, d.h. weniger anwendbare Werte haben (`Survived`, `Pclass`, `Sex`, `Embarked`)  im Gegensatz zu denen, die einen größeren Wertebereich haben (`Age`, `Fare`, `SibSp`, `Parch`)  

Es ist zu beachten dass einige dieser Variablen eine logische Reihenfolge haben (z. B. können die Zahlen auf- oder absteigend sein). 

Variablen mit weniger Werten lassen sich leicht in Säulen/Balken- oder Kreisdiagramme einordnen.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(21, 5))

axes[0].set_title("Embarked")

# Dies zeichnet das Säulendiagramm
df["Embarked"].value_counts().plot.bar(ax=axes[0])

axes[1].set_title("Embarked - Gestapelte")

# alternative Version
dfBar= pd.DataFrame([df["Embarked"].value_counts()])
dfBar.plot.bar(stacked=True, ax=axes[1]);

axes[2].set_title("Passagierklasse")

# Dies zeichnet das Kreisdiagramm
df["Pclass"].value_counts().plot.pie(ax=axes[2])
```

Während andere Verteilungen ein Histogramm, eine Box oder ein Flächendiagramm erfordern (müssen zuerst in numerische Daten umgewandelt werden)

```python
df["Age"] = pd.to_numeric(df["Age"])
df["Fare"] = pd.to_numeric(df["Fare"])

df["SibSp"] = pd.to_numeric(df["SibSp"])
df["Parch"] = pd.to_numeric(df["Parch"])
df["Family"] = df["SibSp"]+df["Parch"]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(21, 5))

axes[0].set_title("Alter")
axes[1].set_title("Fahrpreis")
axes[2].set_title("Größe der Familie")

# Dies zeichnet die Histogramme
df["Age"].plot.hist(alpha=0.5, ax=axes[0])
df["Family"].plot.hist(alpha=0.5, ax=axes[2])

# Dies zeichnet das Boxplot
df["Fare"].plot.box(ax=axes[1])
```

Flächenplots erfordern eine andere Art von Daten

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

base = np.random.rand(10, 4)
colNames = ["a", "b", "c", "d"]

dfAux = pd.DataFrame(base, columns=colNames)

# Dies zeichnet das Flächediagramm
dfAux.plot.area(ax=axes[0], stacked = True)

# Zusätzlich können wir eine Tabelle hinzufügen, um die verwendeten Werte anzuzeigen
axes[1].axis("off")
axes[1].table(cellText=np.round(base, 2), colLabels=colNames, loc='center')
```

Und sie können logisch mit unserem objektiven Merkmal kombiniert werden 

#### **Beispiel**: Frauen und Kinder zuerst!

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

fig.suptitle("Wer hat die Titanic überlebt?")

## Excel-like pivot table für Gender vs Survived
df[["Survived","Sex"]].pivot_table(index='Survived', columns='Sex', aggfunc=len).plot.bar(stacked=True, ax=axes[0])
axes[0].set_title("pro Gender")
axes[0].set_xticklabels(['Nein', 'Ja'], rotation=45, ha='right')

## Histogramm für Alter
axes[1].set_title("pro Alter")

# dies zeichnet die Histogramm
df.groupby('Survived')['Age'].plot.hist(alpha=0.5, legend=True, bins=20, ax=axes[1])

# eine Linie für ein arbiträres Grenzalter von 15 Jahren
axes[1].axvline(x=15, color='black', ls='--', label='RMSE 1')
```

Andere Variablen, die ebenfalls Aussagekräftig sein können

```python
df["Pclass"] = pd.to_numeric(df["Pclass"])
ax = df[["Survived","Pclass"]].pivot_table(index='Pclass', columns='Survived', aggfunc=len).plot.bar(stacked=False, title='Survived vs Passagierklasse')
ax.legend(labels=['Nein','Ja'])
```
