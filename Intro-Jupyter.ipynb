#### **Sample usage**: Reconstruct test data using its matching training data and the full dataset

* **Train set**: https://github.com/datasciencedojo/datasets/blob/master/titanic.csv (_replaced \"\" with \' to match properly all records_)

```python
dfTotal = pd.read_csv('titanicFull.csv')
dfTrain = pd.read_csv('titanicTrain.csv')
print(f"Total records {len(dfTotal)}, Training Set records {len(dfTrain)}")
```

Using `.isin(<collection>)` to check for intersections/complements

```python
trainSetMatches = dfTotal["Name"].isin(dfTrain["Name"])
trainSetMatches.value_counts() # matches must be 891
```

Once the matches reach 100% (891 possible cases), the other values can be saved as Test set. \
If not, further data analysis must be done by slicing to figure out what causes the mismatches. e.g. checking the complement of `dfTotal` in `dfTrain` revealed that the missing cases were due different quotation marks for nicknames

```python
dfTest = dfTotal[~trainSetMatches].copy()
dfTest.to_csv("titanicTest_generated.csv", index=False)
```

#### **Sample usage**: Primer on Feature Analysis

This dataset is often used to model the likelihood of someone surviving the Titanic disaster based on its attributes.  
Let´s review the data we currently have:

The Titanic dataset contains a list of known passengers of the Titanic, with the following variables

* pclass: _Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)_
* survival: _(0 = No; 1 = Yes)_
* sibsp: _Number of Siblings/Spouses Aboard_
* parch: _Number of Parents/Children Aboard_
* embarked: _port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)_
* boat: _id of the lifeboat where the person was counted_
* body: _body identification number_

```python
df = pd.read_csv('titanicFull.csv')
df.info()
```

It has no null values! However we saw previously that instead of nulls, it uses "?" when it has no-data

```python
df = df.replace("?", pd.NA)
df.info()
```

From the variables, we can see that Cabin, Boat, Body and HomeDest are quite empty and may not help further analysis  
While these could be easily removed, note the case of Boat and Body: If someone was associated to a lifeboat or has a body identification number it is easy to suggest their fate.

```python
df2= df[["Body","Boat","Survived"]].copy()
df2.head()
```

These can be interpreted more formally by creating extra variables to determine whether or not the attribute exists  
While this variables can be left to user interpretation, a good convention is to use numbers instead (1 = Yes, 0 = No)

```python
#attribute names can be anything, but they will have to be typed frequently
df2["BodyQ"] = 0
df2.loc[~df2["Body"].isna(), "BodyQ"] = 1

df2["BoatQ"] = 0
df2.loc[~df2["Boat"].isna(),"BoatQ"] = 1
```

Results can be then grouped and tallied

```python
# alternatively df2.value_counts(["BodyQ","Survived"], sort=False)
df2[["BodyQ","Survived"]].groupby(["BodyQ"]).value_counts()
```

```python
# alternatively df2.value_counts(["BoatQ","Survived"], sort=False)
df2[["BoatQ","Survived"]].groupby(["BoatQ"]).value_counts()
```

Having the attributes as numbers also allows to make other calculations such as correlation scores

```python
df2[["BodyQ","BoatQ","Survived"]].corr()
```

These two variables do explain significatively the target attribute. However as the objective is to dig deeper by using ML algorithms and techniques, they are often omitted.  
Note that even by using these variables directly, we will not get exact results but a very reasonable guess.

### Matplotlib

A  library used for graphs and plots. Pandas provides convenience functions to use its features (see https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```python
# Even though most is handled through pandas, some options require it to be loaded directly
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanicFull.csv')
df = df.replace("?", pd.NA)
df.drop(columns=["Cabin","Boat","Body","HomeDest"])
```

The remaining variables can provide insight on the passengers distribution. For this purpose perhaps a graphical approach may be better than just reading lists of numbers.  
We can also identify variables that work as discrete values i.e. have fewer applicable values (`Survived`, `Pclass`, `Sex`, `Embarked`) versus those that use a larger range of values (`Age`, `Fare`, `SibSp`, `Parch`)  
Note that some of these variables have a logical ordering (e.g. Numbers can be ascending or descending) 

Variables with fewer values can be easily classified into Bar or Pie plots

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(21, 5))

axes[0].set_title("Embarked")

# this is what draws the bar plot
df["Embarked"].value_counts().plot.bar(ax=axes[0])

axes[1].set_title("Embarked - Stacked")

# alternative version for the bar plot
dfBar= pd.DataFrame([df["Embarked"].value_counts()])
dfBar.plot.bar(stacked=True, ax=axes[1]);

axes[2].set_title("Passenger Class")

# this is what draws the pie plot
df["Pclass"].value_counts().plot.pie(ax=axes[2])
```

While other distributions may require a Histogram, Box or an Area Plot (need to be converted to numeric data first)

```python
df["Age"] = pd.to_numeric(df["Age"])
df["Fare"] = pd.to_numeric(df["Fare"])

df["SibSp"] = pd.to_numeric(df["SibSp"])
df["Parch"] = pd.to_numeric(df["Parch"])
df["Family"] = df["SibSp"]+df["Parch"]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(21, 5))

axes[0].set_title("Age")
axes[1].set_title("Fare")
axes[2].set_title("Family size")

# this is what draws the histograms
df["Age"].plot.hist(alpha=0.5, ax=axes[0])
df["Family"].plot.hist(alpha=0.5, ax=axes[2])

# this is what draws the box chart
df["Fare"].plot.box(ax=axes[1])
```

Area plots require a different type of data

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

base = np.random.rand(10, 4)
colNames = ["a", "b", "c", "d"]

dfAux = pd.DataFrame(base, columns=colNames)

# this is what draws the area plot
dfAux.plot.area(ax=axes[0], stacked = True)

# additionally we can add a table to show the values used
axes[1].axis("off")
axes[1].table(cellText=np.round(base, 2), colLabels=colNames, loc='center')
```

And they can be logically combined with our objective attribute

#### **Sample usage**: Frauen und Kinder zuerst!

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

fig.suptitle("Who survived the Titanic?")

## Making a Excel-like pivot table for Gender vs Survived
df[["Survived","Sex"]].pivot_table(index='Survived', columns='Sex', aggfunc=len).plot.bar(stacked=True, ax=axes[0])
axes[0].set_title("per Gender")
axes[0].set_xticklabels(['No', 'Yes'], rotation=45, ha='right')

## Using a histogram for Age
axes[1].set_title("per Age")

# this is what draws the historgram
df.groupby('Survived')['Age'].plot.hist(alpha=0.5, legend=True, bins=20, ax=axes[1])

# line for arbitrary cutoff age at 15
axes[1].axvline(x=15, color='black', ls='--', label='RMSE 1')
```

Other variables that could have explanatory power as well

```python
df["Pclass"] = pd.to_numeric(df["Pclass"])
ax = df[["Survived","Pclass"]].pivot_table(index='Pclass', columns='Survived', aggfunc=len).plot.bar(stacked=False, title='Survived')
ax.legend(labels=['No','Yes'])
```
